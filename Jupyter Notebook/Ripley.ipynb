```python
#Importar las bibliotecas requeridas 
import requests
from bs4 import BeautifulSoup
from lxml import etree
import pandas as pd
```

```python
#cargar el contenido de la página web de ripley
r = requests.get('https://simple.ripley.com.pe/tecnologia/tv-y-cine-en-casa/televisores?source=menu&s=mdco')
```

```python
#Verificar estado del código HTTP
r.status_code
```

```python
#crear un objeto BeautifulSoup
soup = BeautifulSoup(r.content,'html.parser')
```

```python
#convertir el objeto de BeautifulSoup en un objeto de etree porque BeautifulSoup por defecto no admite trabajar con XPath
dom = etree.HTML(str(soup))
```

```python
#usar el método .xpath() para buscar elementos.
categorias=dom.xpath('//div[@class="catalog-product-details"]/div[3]/div/div/span')
```

```python
#recorrer categorías y mostrarlo en formato de texto
categorias = [i.text for i in categorias]
categorias
```

```python
#usar el método .xpath() para buscar elementos.
titulos=dom.xpath('//div[@class="catalog-product-details__name"]')
```

```python
#recorrer titulos y mostrarlo en formato de texto
titulos = [i.text for i in titulos]
titulos
```

```python
#recorrer urls y mostrarlo en formato de texto
urls=dom.xpath('//div[@class="catalog-product-item catalog-product-item__container col-xs-6 col-sm-6 col-md-4 col-lg-4"]/a')
urls=[i.get('href') for i in urls]
urls
```

```python
#inspeccionar etiqueta de la página html 
precios=dom.xpath('//li[@title="Precio Internet"]')
```

```python
#recorrer cada uno de los precios y mostrarlo en formato de texto
precios = [i.text for i in precios]
#imprimir desde la posición 2 en adelante y reemplazar el signo "," por un espacio en blacio
precios = [float(i[2:].replace(',', '')) for i in precios]
precios
```

```python
#Verificar que la longitud coincida
print(len(categorias))
print(len(titulos))
print(len(urls))
print(len(precios))
```

```python
#generar dataframe a partir de un diccionario
df=pd.DataFrame({"categoria":categorias,"titulo":titulos, "url":urls,"precio":precios})
df
```

```python
#realizar paginación y obtener el href del primer elemento
siguiente=dom.xpath('//ul[@class="pagination"]/li[5]/a')[0].get('href')
link="https://simple.ripley.com.pe/tecnologia/tv-y-cine-en-casa/televisores"
siguiente=link+siguiente
siguiente
```

```python
#buscar la primera etiqueta de la paginación en formato de texto 
ini = soup.find('a', attrs={"class","is-active"}).text
#convertir de cadena a entero
ini=int(ini)
ini
```

```python
#buscar e imprimir la longitud del rango de paginación
can=len(dom.xpath('//div[@class="catalog-page__footer-pagination"]//li'))-2
can
```

```python
#listas vacías
lista_titulos=[]
lista_url=[]
lista_precios=[]
lista_categorias=[]
```

```python
#comienza el ciclo en la url principal
siguiente='https://simple.ripley.com.pe/tecnologia/tv-y-cine-en-casa/televisores?source=menu&s=mdco'
#recorrer páginas 
while True:
    r=requests.get(siguiente)
    #si el estado es 200: código que se entrega cuando una página web o recurso actúa exactamente como se espera.
    #se ejecuta las líneas de código que realizamos anteriormente
    if r.status_code==200:
        soup = BeautifulSoup(r.content,'html.parser')
        dom = etree.HTML(str(soup))
        #Categorias
        categorias=dom.xpath('//div[@class="catalog-product-details"]/div[3]/div/div/span')        
        categorias = [i.text for i in categorias]
        lista_categorias.extend(categorias) #añadir a la lista
        #Títulos
        titulos=soup.find_all('div',attrs={"class":"catalog-product-details__name"})        
        titulos = [i.text for i in titulos]
        lista_titulos.extend(titulos) #añadir a la lista
        #Url
        urls=dom.xpath('//div[@class="catalog-product-item catalog-product-item__container col-xs-6 col-sm-6 col-md-4 col-lg-4"]/a')
        urls=[i.get('href') for i in urls]
        lista_url.extend(urls) #añadir a la lista
        #Precios
        precios=soup.find_all('li',attrs={"title":"Precio Internet"})        
        precios = [i.text for i in precios]
        precios = [float(i[2:].replace(',', '')) for i in precios]
        lista_precios.extend(precios) #añadir a la lista
        #validamos nuevamente ini y can 
        ini = soup.find('a', attrs={"class","is-active"}).text
        ini=int(ini)
        can=len(dom.xpath('//div[@class="catalog-page__footer-pagination"]//li'))-2
    else:
        break
    #imprimir paginación
    print(ini,can,siguiente)
    #se rompe el bucle cuando la paginación actual sea igual a la paginación total
    if ini==can:
        break
    #retorna a la siguiente página cuando la paginación actual sea diferente a la paginación total
    siguiente=dom.xpath('//ul[@class="pagination"]/li[5]/a')[0].get('href')
    link="https://simple.ripley.com.pe/tecnologia/tv-y-cine-en-casa/televisores"
    siguiente=link+siguiente
```

```python
#validar longitud de las listas (tienen que ser iguales)
print(len(lista_categorias))
print(len(lista_titulos))
print(len(lista_url))
print(len(lista_precios))
```

```python
#convertir a dataframe a partir de un diccionario
df=pd.DataFrame({"categorias":lista_categorias,"titulos":lista_titulos, "url":lista_url,"precio":lista_precios})
df
```

```python
#exportamos a formato csv
df.to_csv('documents/ripley.csv')
```
