```python
#Importar las bibliotecas requeridas 
import pandas as pd
import requests
from bs4 import BeautifulSoup

#descarga de contenidos de la página web
url = 'https://www.transfermarkt.com/1-bundesliga/startseite/wettbewerb/L1/saison_id/2019'
headers={'User-Agent':'Mozilla/5.0'}

#cargar el contenido de la página web de transfermarkt
page = requests.get(url, headers=headers)

#crear un objeto BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')

#inspeccionar las etiquetas de la página html 
table = soup.find('table', attrs={'class':'items'}) #buscar la tabla con la clase'items'
table_body = table.find('tbody') #buscar la tabla con la etiqueta'tbody'
rows = table_body.find_all('tr') #buscar filas de la tabla con la etiqueta'tr'

#crear diccionario vacio
data = []
#recopilación de datos 
for row in rows:
    #Encuentra todos los datos para cada columna
    cols = row.find_all('td')
    #obtener el texto sin formato del elemento
    cols = [ele.text.strip() for ele in cols]
    #agregar al diccionario
    data.append([ele for ele in cols if ele]) 
    
#convertir a dataframe a partir de un diccionario
#agregar nombre a las columnas
df = pd.DataFrame(data, columns=['Club','Squad', 'Edad Promedio de Jugador','N° de Extranjeros', 'Valor de Mercado Promedio de Jugador', 'Valor de Mercado Total'])
#reemplazar el signo "€" por un espacio en blanco y en lugar de "m" y "k" colocar ceros y convertir a un número entero
df['Valor de Mercado Promedio de Jugador'] = df['Valor de Mercado Promedio de Jugador'].str.replace('€', '').str.replace('m', '000000').str.replace('k', '000').str.replace('.', '').astype(int)
#reemplazar el signo "€" y "." por un espacio en blanco y en lugar de "m" colocar ceros y convertir a un número entero 
df['Valor de Mercado Total'] = df['Valor de Mercado Total'].str.replace('€', '').str.replace('m', '000000').str.replace('.', '').astype(float)
df

df.to_csv('documents/transfermarkt.csv')
```
